# Slug Detection HyperParameter Tuning

Hyper parameter tuning was performed to decide on the best values for the following parameters:

Random Forest Hyper Parameters: 

- Number of Trees (n_estimators) 
- Depth of Trees(max_depth) and, 
- Bootstrap (whether columns/rows are dropped per Decision Tree)(bootstrap)

Logistics Regression: 
- Regularization value (C) and,
- Maximum iteration number (max_iter)

Data Engineering Method: 
- Window size (default 5 hours) and 
- Number of splits (default 5) in Feature Vectors

```python
class Slug_Detection(Data):
  def __init__(self, well):
    Data.__init__(self, well)
  
  def jump(self):
    
    # get "jump" column which indicates discontinous data
    self.pd_df["time_diff"] = self.pd_df["ts"].diff() # get time difference between rows
    self.pd_df["jump"] = np.where(self.pd_df["time_diff"] > pd.Timedelta('1 min'), True, False) # check if time difference is larger than 1min

    # group together continuous data with a counter id
    self.pd_df['count_id'] = self.pd_df["jump"].cumsum() 
    
    return
  
  def clean_short_sub(self, min_df_size = 360):
  
    # get number of entries per count_id
    count = self.pd_df.groupby('count_id')['count_id'].count()

    # delete 'clean data' that is less than min_size, default 6 hours
    for i in range(1, len(count)):
      if count[i] < min_df_size:
        self.pd_df = self.pd_df[self.pd_df.count_id != i] # mask out rows where count_id 

    return 
  
  def sub_data(self, **kwargs):
  
    self.sub_df_dict = {}
    
    self.jump() # group clean data in continuous in time groups
    
    #check for kwargs
    min_df_size = 360
    if "min_df_size" in kwargs.keys():
      min_df_size = kwargs["min_df_size"]
    
    self.clean_short_sub(min_df_size=min_df_size) # clean out short continuous data
    self.jump()  #get new, ordered id for groups

    #create a subset dataframe
    for i in range(self.pd_df["count_id"].max()):
      
      temp = pd.DataFrame()
      mask = self.pd_df["count_id"] == i
      temp = self.pd_df[mask].reset_index() # temp is set to clean_df where id is the same, reset indexing 
      
      self.sub_df_dict[i] = temp # save all df to dictionnary

    return 
  
  def slug_check(self, slug_idx, key):
    
    curr_df = self.sub_df_dict[key]
  
    #set it to - 500 to allow for slugs that occur quickly
    first_slug_idx = [-500]

    # variables to check if slugging period has been continuous
    first = False
    last_slug = 0

    # from slug indices, create small lists of 3 consecutive slugs indices
    slug_check = [slug_idx[i:i+3] for i in range(0, len(slug_idx) - 2, 1)]

    #loop through list of 3 indices
    for idx in slug_check: 
      # if gap of 300 minutes since last first slug AND slugs havent been continuous, move to next condition
      if (abs(first_slug_idx[-1] - idx[0]) < 300) | ((abs(last_slug - idx[0]) < 60) & first):
        last_slug = idx[0]
        continue # move to next index
      # if less than 20 minutes between two slug, move to next condition
      elif (abs(idx[0]-idx[1]) > 25) | (abs(idx[1]-idx[0]) > 25):
        continue
      # if less than 2 bar difference between slugs (we want similar flow pattern), move to next condition
      elif (abs(curr_df["WH_P"].iloc[idx[0]] - curr_df["WH_P"].iloc[idx[1]]) > 2.0) | (abs(curr_df["WH_P"].iloc[idx[1]] - curr_df["WH_P"].iloc[idx[2]])> 2.0):
        continue
      else:
        if (len(first_slug_idx) == 1): first = True # first to True, when the first slug of the sub dtaaframe is occuring
        first_slug_idx.append(idx[0]) #first slug is occruing, save index value

    first_slug_idx.pop(0) # loose dummy value set in the first slug index list
    return first_slug_idx
  
  def label_data(self, slug_diff = 3):
    # add a label to each sub data frame

    first_slug = []
    slugs = []

    for key in self.sub_df_dict: #####dict.iteritems(): ?????
      curr_df = self.sub_df_dict[key]

      curr_df["WH_P_diff"] = curr_df["WH_P"].diff()
      curr_df["trend"] = np.where(curr_df["WH_P_diff"] > 0, True, False) # increasing trend, decreasing trend
      curr_df["categories"]= (curr_df["trend"]!=curr_df["trend"].shift()).cumsum() # group by conscutive trend 

      curr_df["WH_P_diff_total"] = curr_df.groupby("categories")["WH_P_diff"].cumsum() # get total cumulative increase and decrease

      # label slug/peaks, when there is a large increse (> slug_diff) followed by a decrease
      curr_df["point_label"] = np.where((curr_df["WH_P_diff_total"] > slug_diff) & (curr_df["WH_P_diff_total"].shift(periods = -1) < 0), True, False)

      # save all slugs occurences indices to list
      slug_index = curr_df.index[curr_df["point_label"] == True].tolist()
      
      # get first slug indices
      first = self.slug_check(slug_index, key)
      
      first_slug.append(first)
      slugs.append(slug_index)

    return first_slug, slugs
  
  def format_data(self, first_slug, size_list=300, max_clean_count = 10000):
    
    self.label = np.array([])
    self.df_list = []
    cnt = 0

    for df, f in zip(self.sub_df_dict.values(),first_slug):

      if not f: # if there's no slug 
        if cnt < max_clean_count:
          
          df = df[:-60] # delete last hour, since it can't be know if the flow would have led to a slug without the data.

          n_splits = int(np.ceil(len(df)/size_list)) # calculate number of splits that can be performed on the sub dataset
          if n_splits > 1:
            overlap = int(np.ceil(((size_list*(n_splits - 1)) - (len(df) - size_list))/(n_splits - 1))) # calculate overlap value
          else:
            overlap = 0

          self.df_list.extend([df[i : i + size_list] for i in range(0, (len(df) - (size_list - overlap)) , (size_list - overlap))]) # extend list of wanted size datasets (default 300)

          for i in range(0, (len(df) - (size_list - overlap)) , (size_list - overlap)):
            self.label = np.append(self.label, [0]) # label each 300_df
            cnt +=1

      else: # if there's slugging in the data
        for first in f:
          if first - size_list >= 0:# if there#s 300 points of data
            self.df_list.append(df[first - size_list: first]) # append list
            self.label = np.append(self.label, [1])# label as slug

    return 
  
  def standardise(self, df):
  
    for feature_ in self.features:
      avg = self.means[feature_][0]
      std = self.stds[feature_][0]
      
      df[feature_] -= avg
      df[feature_] /= std
      
    return df
  
  def feature_vector(self, window_size=5, time_predict=60, percentage_significance=0.1, standardise = True):

    for k in range(len(self.df_list)):
      clean_df = self.df_list[k].copy()
      
      if standardise: # standardise the data
        clean_df = self.standardise(clean_df)
        
      clean_df = clean_df[:-time_predict] #ignore data within prediction time, default 1 hour
      clean_df = clean_df.drop("WH_choke", axis = 1) # choke isnt useful anymore, since must be opened
      self.features = ["WH_P", "DH_P", "WH_T", "DH_T"]

      interval = int(len(clean_df)/window_size) # get window size

      #store features
      clean = []
      if k == 0:
        header = []

      ### loop through all clean slates of data
      #get features from time windows
      for i in range(window_size):
        low = i*interval
        high = (i+1)*interval

        #mean and std
        for f in self.features:
          
          if k == 0:
            header.append(("mean_" + f + "_" + str(i)))
            header.append(("std_" + f + "_" + str(i)))

          clean.append(clean_df[f][low:high].mean())
          clean.append(clean_df[f][low:high].std())

      if k == 0:
        self.X = pd.DataFrame([clean], columns=[*header])
      else:
        dic = dict(zip(header, clean))
        self.X = self.X.append(dic, ignore_index=True)

    
    ### Operations on full new_df, all data
    ### Get Delta, trends and significance of means and std between windows
    for i in range(window_size - 1):
      for f in self.features:
        #get deltas for each feature
        self.X["diff_mean_" + f + "_" + str(i) + "_" + str(i+1)] = self.X["mean_" + f + "_" + str(i+1)] - self.X["mean_" + f + "_" + str(i)]
        self.X["diff_std_" + f + "_" + str(i) + "_" + str(i+1)] = self.X["std_" + f + "_" + str(i+1)] - self.X["std_" + f + "_" + str(i)]

        # if increase then 1, else, 0
        self.X["diff_mean_trend_" + f + "_" + str(i) + "_" + str(i+1)] = np.where(self.X["diff_mean_" + f + "_" + str(i) + "_" + str(i+1)]>0, 1, 0)
        self.X["diff_std_trend_" + f + "_" + str(i) + "_" + str(i+1)] = np.where(self.X["diff_std_" + f + "_" + str(i) + "_" + str(i+1)]>0, 1, 0)

        # if significant 1, else 0
        # significance is a difference of more than percentage_significance, default 10%
        self.X["diff_mean_signif_" + f + "_" + str(i) + "_" + str(i+1)] = np.where(self.X["diff_mean_" + f + "_" + str(i) + "_" + str(i+1)] > self.X[("mean_" + f + "_" + str(i))]*percentage_significance, 1, 0)
        self.X["diff_std_signif_" + f + "_" + str(i) + "_" + str(i+1)] = np.where(self.X["diff_std_" + f + "_" + str(i) + "_" + str(i+1)] > self.X[("std_" + f + "_" + str(i))]*percentage_significance, 1, 0)
    
    for f in self.features:
      binary_mean_diff_col_names = []
      binary_mean_trend_col_names = []
      binary_std_diff_col_names = []
      binary_std_trend_col_names = []
      for i in range(window_size - 1):
        binary_mean_trend_col_names.append("diff_mean_trend_" + f + "_" + str(i) + "_" + str(i+1))
        binary_mean_diff_col_names.append("diff_mean_signif_" + f + "_" + str(i) + "_" + str(i+1))
        binary_std_trend_col_names.append("diff_std_trend_" + f + "_" + str(i) + "_" + str(i+1))
        binary_std_diff_col_names.append("diff_std_signif_" + f + "_" + str(i) + "_" + str(i+1))

      ## number of significant increases
      self.X["num_mean_" + f + "_sign_incr"] = self.X[binary_mean_diff_col_names].sum(axis=1).where(self.X[binary_mean_trend_col_names] == 1, 0)
      self.X["num_std_" + f + "_sign_incr"] = self.X[binary_std_diff_col_names].sum(axis=1).where(self.X[binary_std_trend_col_names] == 1, 0)

      ## number of significant decreases
      self.X["num_mean_" + f + "_sign_decr"] = self.X[binary_mean_diff_col_names].sum(axis=1).where(self.X[binary_mean_trend_col_names] == 0, 0)
      self.X["num_std_" + f + "_sign_decr"] = self.X[binary_std_diff_col_names].sum(axis=1).where(self.X[binary_std_trend_col_names] == 0, 0)

    return 
  
  def split_data(self, test_size=0.7):

    sss = StratifiedShuffleSplit(n_splits=2, test_size=test_size, random_state=0)
    sss.get_n_splits(self.X, self.label)

    train_index, test_index = sss.split(self.X, self.label)

    self.X_train, self.X_test = self.X.iloc[train_index[0]], self.X.iloc[test_index[1]]
    self.y_train, self.y_test = self.label[train_index[0]], self.label[test_index[1]]

    return 

  def RF_train(self, **kwargs):
    
    test_size = 0.7
    if "test_size" in kwargs.keys():
      test_size = kwargs["test_size"]
    self.split_data(test_size=test_size)

    self.rf = RandomForestClassifier()
    self.rf.fit(self.X_train, self.y_train)

    pred_features = self.rf.feature_importances_

    return pred_features
  
  def feature_selection(self, feature_score, top_n=15):
    
    from collections import namedtuple
    
    feature_s = []
    Feature = namedtuple('Feature', 'name score')
    
    for f, s in zip(self.X.columns, feature_score):
      feature_s.append(Feature(f, s))
      
    feature_ranked = sorted(feature_s, key=lambda x: getattr(x, 'score'), reverse=True)
    
    top_feature = []
    for i in range(top_n):
      top_feature.append(feature_ranked[i][0])
    
    return top_feature
  
  
  def RF_predict(self, cf=True, true_label = True, **kwargs):
    
    
    if "X_test" in kwargs.keys():
      self.X_test = kwargs["X_test"]
      
    pred = self.rf.predict(self.X_test)
      
    if true_label:
      
      if "y_test" in kwargs.keys():
        self.y_test = kwargs["y_test"]

      score = self.rf.score(self.X_test, self.y_test)

      cf = confusion_matrix(self.y_test, pred)
    else:
      score, cf = 0, 0
    
    return pred, score, cf
  
  def LogReg_train(self, top_features):
    
    if top_features:
      self.logreg_features = top_features
      # update feature vector
      self.X_train = self.X_train[[*self.logreg_features]]

    self.log = LogisticRegression()
    
    self.log.fit(self.X_train, self.y_train)
    
    return 
  
  def LogReg_pred(self, true_label=True, **kwargs):
    
    if "X_test" in kwargs.keys():
      self.X_test = kwargs["X_test"]
      
    if self.logreg_features:
      self.X_test = self.X_test[[*self.logreg_features]]
    
    log_pred = self.log.predict(self.X_test)
    
    if true_label:
      
      if "y_test" in kwargs.keys():
        self.y_test = kwargs["y_test"]
    
      proba = self.log.predict_proba(self.X_test)

      score = self.log.score(self.X_test, self.y_test)

      cf = confusion_matrix(self.y_test, pred)
    
    else:
      score, cf = 0, 0
    
    return log_pred, proba, score, cf
    
    
  def data_prep(self, **kwargs):
    
    # check pandas dataframe exists
    assert(not self.pd_df.empty)
    
    # split dataframe into smaller continous dataframes
    self.sub_data(**kwargs)
    
   # check for kwargs
    slug_diff=3
    if "slug_diff" in kwargs.keys(): # if slug definiton changes
      slug_diff = kwargs["slug_diff"]    
    
     # get first slug indexes
    first_idx, slug_idx = self.label_data(slug_diff = slug_diff)
    
    # check for kwargs
    size_list=300
    if "size_list" in kwargs.keys():
      size_list = kwargs["size_list"]
    max_clean_count = 10000
    if "max_clean_count" in kwargs.keys():
      max_clean_count = kwargs["max_clean_count"]
    
    # format data into size_list (default 300) long dataframes, with label list
    self.format_data(first_idx, size_list=size_list, max_clean_count = max_clean_count)
    
    #check for kwargs
    window_size=5
    if "window_size" in kwargs.keys():
      window_size = kwargs["window_size"]
    time_predict=60
    if "time_predict" in kwargs.keys():
      time_predict = kwargs["time_predict"]
    percentage_significance=0.1
    if "percentage_significance" in kwargs.keys():
      percentage_significance = kwargs["percentage_significance"]
    standardise=True
    if "standardise" in kwargs.keys():
      standardise = kwargs["standardise"]
    
    # create data feature vectors
    self.feature_vector(window_size=window_size, time_predict=time_predict, percentage_significance=percentage_significance, standardise=standardise)
    
    return
  
  def plot_raw_slugs(self, variables=["WH_P", "DH_P"], scaled=True, n_examples = 10, first_sample = 10):
    
    assert(self.sub_df_dict)
    
    x_list = [x for x in range(first_sample, first_sample+n_examples)]
    
    f, s = self.label_data()
    
    fig, ax = plt.subplots(n_examples, 1, figsize= (30, 30))
    plt.tight_layout()

    for i, x in enumerate(x_list):
      
      for v in variables:
        if (v != "WH_P") & (scaled):
          ax[i].plot(self.sub_df_dict[x]["ts"], self.sub_df_dict[x][v] - self.sub_df_dict[x][v][0] + 20, "-", label = str(v))
        else:
          ax[i].plot(self.sub_df_dict[x]["ts"], self.sub_df_dict[x][v], "-", label = str(v))
          
      ax[i].plot(self.sub_df_dict[x]["ts"][s[x]], self.sub_df_dict[x]["WH_P"][s[x]], "ro", label = "Slug peaks")
      ax[i].plot(self.sub_df_dict[x]["ts"][f[x]], self.sub_df_dict[x]["WH_P"][f[x]], "m*", markersize = 20, label = "First slug")
      ax[i].plot(self.sub_df_dict[x]["ts"][f[x]]-pd.Timedelta('5 h'), self.sub_df_dict[x]["WH_P"][f[x]], "c*", markersize = 20, label = "Start interval")
      ax[i].set_xlabel("Time")
      ax[i].grid(True, which='both')
      ax[i].set_ylabel("Pressure in barG")
      ax[i].legend()
      
    display(fig)
  
  def plot_X(self, start = 60, variables=["WH_P", "DH_P"], scaled = True, n_examples = 3):
    
    fig, ax = plt.subplots(3,2, figsize= (15, 7))
    plt.tight_layout()
    
    data1, data2 = True, True
    plot1, plot2 = 0, 0
    for idx, val in enumerate(sd.label):
      if idx > start:
        if (val == 1) & (data1):
          for v in variables:
            if (v != "WH_P") & (scaled):
              ax[plot1][0].plot(self.df_list[idx]["ts"], self.df_list[idx][v] - self.df_list[idx][v].iloc[0] + 20, "-", label = str(v))
            else:
              ax[plot1][0].plot(self.df_list[idx]["ts"], self.df_list[idx][v], "-", label = str(v))
          ax[plot1][0].set_xlabel("Time")
          ax[plot1][0].grid(True, which='both')
          ax[plot1][0].set_ylabel("Pressure in barG")
          ax[0][0].set_title("Interval leading to slug flow")
          ax[plot1][0].legend()
          plot1 +=1
          if plot1 == n_examples:
            data1 = False
          
        elif (val == 0) & (data2):
          for v in variables:
            if (v != "WH_P") & (scaled):
              ax[plot2][1].plot(self.df_list[idx]["ts"], self.df_list[idx][v] - self.df_list[idx][v].iloc[0] + 20, "-", label = str(v))
            else:
              ax[plot2][1].plot(self.df_list[idx]["ts"], self.df_list[idx][v], "-", label = str(v))
          ax[plot2][1].set_xlabel("Time")
          ax[plot2][1].grid(True, which='both')
          ax[plot2][1].set_ylabel("Pressure in barG")
          ax[0][1].set_title("Interval leading to normal flow")
          ax[plot2][1].legend()
          plot2 +=1
          if plot2 == n_examples:
            data2 = False

    display(fig)
    
  def plot_cm(self, cm, title='Confusion matrix'):
    
    # from matplotlib
    fig, ax = plt.subplots()
    
    im = ax.imshow(cm, interpolation='nearest', cmap='RdYlGn')
    ax.figure.colorbar(im, ax=ax)
  
    ax.set(xticks=np.arange(cm.shape[1]),
           yticks=np.arange(cm.shape[0]),
           xticklabels=["Normal", "Slug Flow"], yticklabels=["Normal", "Slug Flow"],
           title=title,
           ylabel='True label',
           xlabel='Predicted label')

    # Loop over data dimensions and create text annotations.
    fmt = '.2f'
    thresh = cm.max() / 2.
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax.text(j, i, format(cm[i, j], fmt),
                    ha="center", va="center",
                    color="white" if cm[i, j] > thresh else "black")
    fig.tight_layout()
    display(fig)

    
    
    
```

## Hyperparameter Tuning Methodology

```python
hyper = Slug_Detection(data_spark)
hyper.timeframe(start="01-JAN-14 09:09", end="30-SEP-18 09:09")
hyper.data_range()
hyper.clean_choke(method = "99")
pd_df = hyper.df_toPandas()
```

```python

size_list = [240, 300, 360]
window_size = [4, 5, 6]

scores = []
for sl in size_list:
  for ws in window_size:
    hyper.data_prep(size_list=sl, window_size=ws, standardise=False)
    pred_f = hyper.RF_train() 
    rf_pred, rf_score, rf_cm = hyper.RF_predict(true_label = True) 
    scores.append(rf_score)
    print("Window size:", sl, "Number of Splits:", ws, "% Score:",  round(rf_score*100, 3))

# fig, ax = plt.subplots(1, 1)
# ax.plot()
```

Overall, by performing this search grid couple of times, the results seem to indicate that 240 and 300 points long interval performs better than a 360 one. For practical reson, 240 minutes interval is chosen over 300. With a shorter interval, more slug flow data is likely to be sourced from the raw data, as the requirements (shorter continous set) are relaxed.

The number of splits in the continous data is not conclusive, and will therefore be kept at 5 splits.
